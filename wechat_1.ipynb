```python
import numpy as np
import pandas as pd
import csv
```

```python
#funciton of reading csv file
def read_csv(in_file):
    out_file = []
    with open(in_file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        next(csv_reader, None)
        for line in csv_reader:
            out_file.append([line[0], line[1], line[3], line[4], line[5], line[6], line[7]])
    out_file = pd.DataFrame(out_file)
    return out_file
```

```python
#replace software name accroding to app list file_app
def app_uid_avg(df):
    for i,v in file_app.iterrows():
        df.loc[df['app_id'] == v[0], 'app_softname'] = v[1]
    df['vv'] = df['vv'].apply(pd.to_numeric, errors='coerce')
    df['days'] = df['days'].apply(pd.to_numeric, errors='coerce')
    df['usetime'] = df['usetime'].apply(pd.to_numeric, errors='coerce')
    return df
```

```python
from datetime import datetime
df_inapp['begin_time'] = pd.to_datetime(df_inapp['begin_time'], unit = 's')
df_inapp['end_time'] = pd.to_datetime(df_inapp['end_time'], unit = 's')
```

```python
df_inapp_times_app = df_inapp.groupby(['content_x'])\
    .agg({'dif_time':lambda x: x.count(),'userid': lambda x: x.nunique()}).reset_index()
df_inapp_times_app['app_times'] = round(df_inapp_times_app['dif_time']/df_inapp_times_app['userid'], 1)
df_inapp_times_app.sort_values(['content_x','app_times'], ascending = [True, False])\
    .groupby('content_x').head(3)
```
